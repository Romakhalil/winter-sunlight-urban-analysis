```python
import pandas as pd
import pvlib
```

```python
latitude = 52.5200
longitude = 13.4050

location = pvlib.location.Location(latitude, longitude, tz="Europe/Berlin")
location
```

```python
time = pd.DatetimeIndex(
    ["2025-12-21 12:00"],
    tz="Europe/Berlin"
)

time
```

```python
sun_pos = location.get_solarposition(time)

sun_pos
```

```python
altitude = sun_pos["apparent_elevation"].iloc[0]
azimuth = sun_pos["azimuth"].iloc[0]

altitude, azimuth
```
